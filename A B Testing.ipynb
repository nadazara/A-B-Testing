```python
data_1.columns = ['campaign name' , 'date' , 'sales', "number of impressions" ,'reach' , 'number of clicks' , 'number of searches', 'number of view content' , 'number of add to cart' , 'number of purchase' ]
data_2.columns = ['campaign name' , 'date' , 'sales', "number of impressions" ,'reach' , 'number of clicks' , 'number of searches', 'number of view content' , 'number of add to cart' , 'number of purchase' ]
```

```python
print(data_1 .columns)
print(data_2.columns)
```

```python
print(data_1.isnull().sum())
```

```python
print(data_2.isnull().sum())
```

```python
# replace null values with mean
data_1['reach'].fillna(data_1['reach'].mean(), inplace = True)
data_1['number of impressions'].fillna(data_1['number of impressions'].mean(), inplace = True)
```

```python
data_1['number of clicks'].fillna(data_1['number of clicks'].mean(), inplace = True)
data_1['number of searches'].fillna(data_1['number of searches'].mean(), inplace = True)
data_1['number of add to cart'].fillna(data_1['number of add to cart'].mean(), inplace = True)
data_1['number of view content'].fillna(data_1['number of view content'].mean(), inplace = True)
data_1['number of purchase'].fillna(data_1['number of purchase'].mean(), inplace = True)
```

```python
data_1['date'] = pd.to_datetime(data_1['date'],format='%d.%m.%Y')
data_2['date'] = pd.to_datetime(data_2['date'],format='%d.%m.%Y')
```
